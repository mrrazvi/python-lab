```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
rents = pd.read_csv('../../data/dubai-real-estate-rents-2025-12.csv')
```

```python
rents.head()
```

```python
prop_sub_types = rents['PROP_SUB_TYPE_EN'].value_counts().head(10)
```

```python
prop_sub_types
```

```python
plt.barh(prop_sub_types.index, prop_sub_types.values)
plt.title('Sales Transactions Count')
plt.xlabel('Count')
plt.ylabel('Property Types')
plt.show()
```
